# Proyecto Final: Identificación de Clientes en Riesgo de Abandono en Gimnasios

## Introducción

Este proyecto tiene como objetivo desarrollar un modelo de clasificación que permita identificar a los clientes de un gimnasio en riesgo de abandonar, utilizando datos sobre sus hábitos de ejercicio, características físicas y métricas fisiológicas. La clasificación de los clientes se basa en su frecuencia de asistencia semanal, dividiéndolos en tres categorías: **baja**, **media** y **alta**.

El modelo se enfoca en proporcionar insights valiosos para mejorar la retención de los clientes y personalizar las estrategias de motivación, impactando directamente en la experiencia del cliente y en los ingresos del gimnasio.

---

## Contexto

En la industria del fitness, mantener la fidelidad de los clientes es un desafío constante. A menudo, los gimnasios enfrentan el problema del abandono de sus miembros debido a la falta de motivación, progreso lento o falta de compromiso. Este proyecto busca resolver este problema mediante un análisis estadístico y un modelo de clasificación basado en datos reales.

El dataset utilizado contiene información detallada de los miembros del gimnasio, incluyendo variables como:
- **Datos demográficos:** Edad y género.
- **Hábitos de ejercicio:** Frecuencia de asistencia, duración de la sesión, y calorías quemadas.
- **Métricas fisiológicas:** Ritmo cardíaco promedio y en reposo.
- **Características físicas:** Índice de masa corporal (BMI) y porcentaje de grasa corporal.

---

## Objetivos

1. **Clasificación:** Desarrollar un modelo de clasificación para segmentar a los clientes en categorías de frecuencia de asistencia (baja, media, alta).
2. **Análisis estadístico:** Explorar las características y patrones de las variables predictoras.
3. **Generar valor:** Identificar a los clientes en riesgo de abandonar y proponer estrategias de retención basadas en los resultados.

---

## Justificación

Un modelo como este genera valor significativo para los gimnasios al:
- **Mejorar la retención de clientes:** Identificando patrones tempranos de abandono.
- **Diseñar estrategias personalizadas:** Basadas en el nivel de compromiso de los clientes.
- **Optimizar recursos:** Enfocando los esfuerzos en las áreas y clientes con mayor necesidad de intervención.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Cargar el dataset
dataset_path = 'dataset.csv'
data = pd.read_csv(dataset_path)

# Mostrar las primeras filas para inspección inicial
data.head()
```

# Análisis Estadístico Profundo de las Variables del Modelo

En este análisis, se evaluarán las métricas estadísticas clave de las variables involucradas en el modelo de clasificación (`Age`, `BMI`, `Avg_BPM`, `Resting_BPM`, `Session_Duration`, `Calories_Burned`), proporcionando su respectiva interpretación y gráficos descriptivos.

## 1. Variables Numéricas

### **1.1 Estadísticas Descriptivas**
El análisis incluye:
- **Media (mean):** Promedio de los valores.
- **Desviación Estándar (std):** Mide la dispersión de los datos.
- **Varianza (variance):** Variabilidad de los datos.
- **Asimetría (skewness):** Indica si la distribución está sesgada.
- **Curtosis (kurtosis):** Mide si los datos tienen colas más largas o cortas que una distribución normal.


```python
# Variables seleccionadas para el análisis (numéricas y categóricas relevantes)
num_variables = ['Age', 'BMI', 'Avg_BPM', 'Resting_BPM', 'Session_Duration (hours)', 'Calories_Burned']
cat_variables = ['Gender', 'Frequency_Category']

# Cálculo de métricas estadísticas
num_stats = data[num_variables].describe().transpose()
num_stats['variance'] = data[num_variables].var()
num_stats['skewness'] = data[num_variables].skew()
num_stats['kurtosis'] = data[num_variables].kurtosis()

# Mostrar estadísticas descriptivas
print(num_stats)
```

```python
# Graficar distribuciones de las variables numéricas
plt.figure(figsize=(15, 10))
for i, var in enumerate(num_variables, 1):
    plt.subplot(2, 3, i)
    sns.histplot(data[var], kde=True, bins=20, color='blue')
    plt.title(f'Distribución de {var}')
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()
```

```python
# Categorizar 'Workout_Frequency (days/week)' en tres clases: baja, media y alta
def categorize_frequency(freq):
    if freq <= 2:
        return "Baja"
    elif 2 < freq <= 4:
        return "Media"
    else:
        return "Alta"

data['Frequency_Category'] = data['Workout_Frequency (days/week)'].apply(categorize_frequency)


# Tablas de frecuencia
cat_analysis = {var: data[var].value_counts() for var in cat_variables}
for var, counts in cat_analysis.items():
    print(f"Frecuencia de {var}:\n{counts}\n")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Dividir datos
X = data[['Age', 'BMI', 'Avg_BPM', 'Resting_BPM', 'Session_Duration (hours)', 'Calories_Burned']]
y = data['Frequency_Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar modelo
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluación
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}
grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3)
grid_search.fit(X_train, y_train)
```

```python
feature_importances = model.feature_importances_
print(sorted(zip(feature_importances, X.columns), reverse=True))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV

# Aplicar SMOTE para balancear las clases en los datos de entrenamiento
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Configuración de GridSearch para optimizar hiperparámetros de Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring='accuracy',
    n_jobs=-1
)

# Entrenar el modelo con los datos balanceados
grid_search.fit(X_resampled, y_resampled)

# Mejor combinación de hiperparámetros
best_params = grid_search.best_params_

# Evaluar el modelo optimizado en el conjunto de prueba
best_model = grid_search.best_estimator_
y_pred_optimized = best_model.predict(X_test)

# Resultados de clasificación y matriz de confusión
classification_results_optimized = classification_report(y_test, y_pred_optimized)
conf_matrix_optimized = confusion_matrix(y_test, y_pred_optimized)

print(best_params)
print(classification_results_optimized)
print(conf_matrix_optimized)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Crear y entrenar un modelo Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42, n_estimators=300, max_depth=20, min_samples_split=2)
gb_model.fit(X_resampled, y_resampled)

# Predicción en los datos de prueba
y_pred_gb = gb_model.predict(X_test)

# Evaluación del modelo Gradient Boosting
classification_results_gb = classification_report(y_test, y_pred_gb)
conf_matrix_gb = confusion_matrix(y_test, y_pred_gb)

print(classification_results_gb)
print(conf_matrix_gb)
```

### **Recomendación Final:**
Utilizar **Random Forest Base** si el enfoque está en resultados generales. Implementar **Gradient Boosting** en escenarios donde la equidad entre clases sea más importante.
