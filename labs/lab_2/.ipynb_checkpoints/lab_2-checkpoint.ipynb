## Parte 1: Numpy

En esta primera parte, trabajaremos con **Numpy** para realizar diversas operaciones con arrays. La manipulación de arrays es esencial para el manejo de datos en econometría y análisis predictivo.

```python
import numpy as np
```

### Problema 1: Tamaño en Bytes de un Array de Booleanos

**Enunciado:**
¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

#### Solución

```python
# Crear un array de 1000 elementos booleanos
array_booleano = np.ones(1000, dtype=bool)

# Calcular el tamaño en bytes
tamaño_bytes = array_booleano.nbytes
print(f"El tamaño en bytes de un array de 1000 elementos booleanos es: {tamaño_bytes} bytes")
```

### Problema 2: Array de Enteros Pares en Orden Decreciente

**Enunciado:**
Cree un array cuyos elementos sean los enteros pares en el rango [1, 100] y en orden decreciente. Muestre los 10 últimos por pantalla.

#### Solución

```python
# Crear un array con los enteros pares en el rango [1, 100] en orden decreciente
array_pares_decreciente = np.arange(100, 0, -2)

# Mostrar los últimos 10 elementos
print("Los últimos 10 elementos del array son:", array_pares_decreciente[-10:])
```

### Problema 3: Selección de Subarrays

**Enunciado:**
Dado el array:

$$
\begin{matrix}
  2 & 4 & 5 & 6 \\
  0 & 3 & 7 & 4 \\
  8 & 8 & 5 & 2 \\
  1 & 5 & 6 & 1
\end{matrix}
$$

Seleccione con una instrucción el subarray de elementos:

$$
\begin{matrix}
  0 & 3 & 7 & 4 
\end{matrix}
$$

Después, seleccione el subarray de elementos:

$$
\begin{matrix}
  2 & 5 \\
  8 & 5
\end{matrix}
$$


```python
# Crear el array dado
array = np.array([
    [2, 4, 5, 6],
    [0, 3, 7, 4],
    [8, 8, 5, 2],
    [1, 5, 6, 1]
])

# Seleccionar el subarray [0, 3, 7, 4] (segunda fila completa)
subarray_1 = array[1, :]  # Segunda fila completa
print("Subarray 1:\n", subarray_1)

# Seleccionar el subarray [[2, 5], [8, 5]] (elementos en filas 0 y 2, columnas 0 y 2)
subarray_2 = array[[0, 2], :][:, [0, 2]]
print("Subarray 2:\n", subarray_2)
```

### Problema 4: Media Aritmética de la Diferencia entre dos Arrays

**Enunciado:**
Dados los arrays `a = [1, 4, 2, 7]` y `b = [1, 3, 2, 9]`, obtenga la media aritmética de la diferencia (a - b).

#### Solución


```python
# Definir los arrays a y b
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# Calcular la diferencia (a - b)
diferencia = a - b

# Calcular la media aritmética de la diferencia
media_aritmetica = np.mean(diferencia)
print("La media aritmética de la diferencia (a - b) es:", media_aritmetica)
```

### Problema 5: Agregar Coordenadas a un Array y Mostrar Dimensiones

**Enunciado:**
Dado el array con coordenadas (x, y):

$$
\begin{matrix}
  1.33 & 4.5 \\
  30.0 & 10.7 \\
  70.2 & 0.5
\end{matrix}
$$


1. Agregue a este array las coordenadas (37.1, -3.6).
2. Muestre en pantalla las dimensiones del nuevo array.

#### Solución


```python
# Crear el array inicial con coordenadas (x, y)
array_coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5]
])

# Agregar las coordenadas (37.1, -3.6)
nuevas_coordenadas = np.array([[37.1, -3.6]])
array_coordenadas = np.vstack((array_coordenadas, nuevas_coordenadas))

print("Nuevo array:\n", array_coordenadas)

# Mostrar las dimensiones del nuevo array
print("\nDimensiones del nuevo array:", array_coordenadas.shape)
```

### Problema 6: Transponer y Agregar Nuevas Coordenadas a un Array

**Enunciado:**
1. Copie el array del ejercicio anterior y transpóngalo.
2. Agregue ahora dos nuevos pares de coordenadas: (10.8, 3.0) y (35.8, 12.0).
3. Muestre en pantalla las dimensiones del array resultante.

#### Solución

```python
# Crear el array inicial con coordenadas (x, y)
array_coordenadas = np.array([
    [1.33, 4.5],
    [30.0, 10.7],
    [70.2, 0.5],
    [37.1, -3.6]
])

# Transponer el array
array_transpuesto = array_coordenadas.T

# Crear un array con las nuevas coordenadas y agregarlo al array transpuesto
nuevas_coordenadas = np.array([[10.8, 3.0], [35.8, 12.0]])
array_transpuesto = np.hstack((array_transpuesto, nuevas_coordenadas))

print("Nuevo array:\n", array_transpuesto)

# Mostrar las dimensiones del array resultante
print("\nDimensiones del array resultante:", array_transpuesto.shape)
```

## Parte 2: Exploración y Minería de Datos con Pandas y Matplotlib

En esta parte del laboratorio, utilizaremos **Pandas** y **Matplotlib** para realizar una serie de análisis y visualizaciones de datos sobre el dataset `NucleosPoblacion.csv`. Este archivo contiene información sobre municipios, incluyendo sus poblaciones y coordenadas geográficas.

### Instrucciones

Asegúrate de haber cargado el dataset correctamente y de revisar su estructura antes de proceder con los ejercicios. Cada ejercicio debe resolverse en una celda de código con los comentarios necesarios para explicar cada paso.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset
nucleos_poblacion = pd.read_csv('NucleosPoblacion.csv')

# Display the structure of the dataset: columns and data types
nucleos_poblacion.info()

# Mostrar las primeras filas del dataset para revisar su estructura
nucleos_poblacion.head()
```

### Problema 1: Municipios con Más de 100,000 Habitantes

**Enunciado:** ¿Cuántos municipios tienen más de 100,000 habitantes?

#### Solución

```python
# Filtrar municipios con más de 100,000 habitantes
municipios_100k = nucleos_poblacion[nucleos_poblacion['Poblacion'] > 100000]

# Contar el número de municipios
num_municipios_100k = municipios_100k.shape[0]
print("Número de municipios con más de 100,000 habitantes:", num_municipios_100k)
```

### Problema 2: Gráfica de Barras de Población por Ciudad

**Enunciado:**
Usando el dataset `NucleosPoblacion.csv`:
1. Realiza una gráfica de barras sobre la población de cada ciudad, ordenada de menor a mayor.
2. Responde:
   - a) ¿Cuál es la segunda ciudad más poblada?
   - b) ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

#### Solución

```python
# Ordenar los datos por población de mayor a menor
nucleos_ordenados = nucleos_poblacion.sort_values(by='Poblacion', ascending=False).reset_index(drop=True)

# Ajustar el tamaño de la figura y el tamaño de la fuente para evitar el traslape de etiquetas
plt.figure(figsize=(12, 15))
plt.barh(nucleos_ordenados['Municipio'], nucleos_ordenados['Poblacion'])
plt.xlabel('Población')
plt.ylabel('Municipio')
plt.title('Población por Ciudad (Ordenada de Menor a Mayor)')
plt.gca()  # Invertir el eje Y para que la ciudad más poblada esté en la parte superior
plt.xticks(rotation=45)  # Rotar etiquetas en el eje X para mejorar legibilidad
plt.yticks(rotation=45)
plt.tight_layout()  # Ajustar diseño para evitar que las etiquetas se recorten
plt.show()

# Obtener la segunda ciudad más poblada
ciudad_segunda_mas_poblada = nucleos_ordenados.iloc[1]['Municipio']
print("a) La segunda ciudad más poblada es:", ciudad_segunda_mas_poblada)

# Encontrar la posición de Granada en el ranking
posicion_granada = nucleos_ordenados[nucleos_ordenados['Municipio'] == 'Granada'].index[0] + 1
print("b) Granada ocupa la posición:", posicion_granada)
```

### Problema 3: Municipios de Tarragona con Más de 32,000 Habitantes

**Enunciado:**
¿Cuántos municipios de la provincia de Tarragona tienen más de 32,000 habitantes?

#### Solución

```python
# Filtrar municipios de la provincia de Tarragona con más de 32,000 habitantes
municipios_tarragona_32k = nucleos_poblacion[(nucleos_poblacion['Provincia'] == 'Tarragona') & 
                                             (nucleos_poblacion['Poblacion'] > 32000)]

# Contar el número de municipios que cumplen el criterio
num_municipios_tarragona_32k = municipios_tarragona_32k.shape[0]
print("Número de municipios en Tarragona con más de 32,000 habitantes:", num_municipios_tarragona_32k)
```

### Problema 4: Municipio Más al Norte

**Enunciado:** ¿Cuál es el municipio situado más al norte? (Usa el valor de la coordenada "Y" que representa la latitud en grados).
- Proporciona también la provincia a la que pertenece y su población.

#### Solución


```python
# Encontrar el municipio con la mayor coordenada "Y" (más al norte)
municipio_mas_norte = nucleos_poblacion.loc[nucleos_poblacion['Y'].idxmax()]

# Mostrar el municipio más al norte, su provincia y su población
print("El municipio más al norte es:", municipio_mas_norte['Municipio'])
print("Provincia:", municipio_mas_norte['Provincia'])
print("Población:", municipio_mas_norte['Poblacion'])
print("Coordenada Y (Latitud):", municipio_mas_norte['Y'])
```

### Problema 5: Estadísticas Descriptivas de la Población en Granada

**Enunciado:** Obtén la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

#### Solución

```python
# Filtrar los municipios de la provincia de Granada
municipios_granada = nucleos_poblacion[nucleos_poblacion['Provincia'] == 'Granada']

# Calcular las estadísticas descriptivas
media_poblacion = municipios_granada['Poblacion'].mean()
mediana_poblacion = municipios_granada['Poblacion'].median()
desviacion_std_poblacion = municipios_granada['Poblacion'].std()
max_poblacion = municipios_granada['Poblacion'].max()
min_poblacion = municipios_granada['Poblacion'].min()

# Mostrar los resultados
print("Estadísticas de la población de los municipios en Granada:")
print("Media:", media_poblacion)
print("Mediana:", mediana_poblacion)
print("Desviación estándar:", desviacion_std_poblacion)
print("Valor máximo:", max_poblacion)
print("Valor mínimo:", min_poblacion)
```

### Problema 6: Histograma de la Población por Provincia

**Enunciado:** 
Realiza un histograma con la población de los municipios para cada una de las provincias.

#### Solución

```python
# Obtener las provincias únicas
provincias = nucleos_poblacion['Provincia'].unique()

# Crear histogramas de la población para cada provincia
for provincia in provincias:
    # Filtrar los municipios de la provincia actual
    municipios_provincia = nucleos_poblacion[nucleos_poblacion['Provincia'] == provincia]
    
    # Crear el histograma
    plt.figure(figsize=(8, 6))
    plt.hist(municipios_provincia['Poblacion'], bins=20, edgecolor='black')
    plt.title(f'Histograma de Población - Provincia de {provincia}')
    plt.xlabel('Población')
    plt.ylabel('Frecuencia')
    plt.show()
```
